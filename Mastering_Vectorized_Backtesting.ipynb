### Trading strategies

* ***Simple moving averages (SMA):***
A signal is derived, for example, when an SMA defined on a shorter time window - say 42 days - crosses an SMA defined on a longer time window - say 252 days.


* ***Momentum strategy:***
It is based on the hypothesis that recent performance will persist for some additional time.


* ***Mean-reversion strategy:***
It assumes that stock prices or prices of other financial instruments tend to revert to some mean level or to some trend level when they have deviated too much from such levels.

##### Generate signals

***Rule:*** "Go long whenever the shorter SMA is above the longer one and vice versa."

Long position = 1          Short position = -1

```python
data['position'] = np.where(data['SMA1'] > data['SMA2'], 1, -1)
```

```python
data.dropna(inplace=True)
data.head(5)
```

```python
data['position'].plot(ylim=[-1.1,1.1],
                     title='Market Position',
                     figsize=(10,6));
```

```python
data['returns'] = np.log(data['price'] / data['price'].shift(1))
data.tail(5)
```

```python
# Histogramas de rendimientos

data.hist(column='returns', bins=35, figsize=(10,6))
```

To derive the ***strategy returns***, multiply the position column --shifted by one trading day-- with the returns column.

Since log returns are additive, calculating the sum over the columns returns and strategy provides a first comparison of the performance of the strategy relative to the base investment itself.

```python
data['strategy'] = data['position'].shift(1) * data['returns']
data.tail(5)
```

```python
# Sums up the single log return values for both the swtock and the strategy

data[['returns', 'strategy']].sum()
```

```python
# Applies the exponential function to the sum of the log returns to calculate the gross performance

data[['returns', 'strategy']].sum().apply(np.exp)
```

Calculating the cumulative sum over time with cumsum and, based on this, the cumulative returns by applying the exponential function np.exp() gives a more comprenhensive picture of how the strategy compares to the performance of the base financial instrument over time.


```python
data[['returns', 'strategy']].cumsum().apply(np.exp).plot(figsize=(10,6))
```

##### Other risck statistics often of interest in the context of trading strategy performances are the maximum drawdown and the longest drawdown period.

A helper statistic to use in this context is the cumulative maximum gross performance as calculated by the cummax() method applied to the gross performance of the strategy.

```python
# Defines a new column, cmret, with the gross performance over time

data['cumret'] = data['strategy'].cumsum().apply(np.exp)
```

```python
# Defines yet another column with the running maximum value of the gross performance

data['cummax'] = data['cumret'].cummax()
```

```python
# Plots the two new columns of the DataFrame object

data[['cumret', 'cummax']].dropna().plot(title='Gross performance and cum max perf. SMA-Strategy', figsize=(10,6))
```

```python
# Where are the differences equal to zero?

drawdown = data['cummax'] - data['cumret']

# Picks out the maximum value from all the differences
max(drawdown)
```

#### Strategies based on Momentum

There are two basic types of momentum strategies:

* ***Cross-sectional strategy***: Buy those instruments that have recently outperformed relative to their peers (or a benchmark) and sell those instruments that have underperformed. 


*  ***Time series strategy***: Buy those instruments that have recently performed well and sell those instruments that have recently performed poorly using the past returns of the instrument itself as a benchmark.

```python
data = pd.DataFrame(raw['XAU='])
data
```

```python
data.rename(columns={'XAU=':'price'}, inplace=True)
data
```

```python
data['returns'] = np.log(data['price'] / data['price'].shift(1))
data
```

```python
data['position'] = np.sign(data['returns'])
data
```

```python
data['strategy'] = data['position'].shift(1) * data['returns']
data
```

```python
data[['returns', 'strategy']].dropna().cumsum().apply(np.exp).plot(figsize=(10,6))
```

```python
period = 3
data['position'] = np.sign(data['returns'].rolling(period).mean())
data['strategy'] = data['position'].shift(1) * data['returns']
data[['returns', 'strategy']].dropna().cumsum().apply(np.exp).plot(figsize=(10,6))
```
